```python
df.drop(['Phone','Account Length'],axis=1,inplace=True)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
v=['State','Area Code','International Plan','VMail Plan']
for i in v:
    df[i] = le.fit_transform(df[i])
```

```python
#from sklearn.preprocessing import StandardScaler
#sc=StandardScaler()
```

```python
x=df.drop('Churn',axis=1)
```

```python
#x=pd.DataFrame(x,columns=['State','Area Code','International Plan','VMail Plan','VMail Message','Day Mins','Day Calls','Day Charge','Eve Mins','Eve Calls','Eve Charge','Night Mins','Night Calls','Night Charge','International Mins','International calls','International Charge','CustServ Calls'])
```

```python
x.head()
```

```python
y=df['Churn']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
predictions=lr.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
print(accuracy_score(y_test,predictions))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier,AdaBoostClassifier,gradient_boosting
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import cross_val_score
```

```python
df.head()
```

```python
mns=[]
stds=[]
clfs=[KNeighborsClassifier(),SVC(),DecisionTreeClassifier(),RandomForestClassifier(),ExtraTreesClassifier(),AdaBoostClassifier(),MultinomialNB()]
for i in clfs:
    cvs=cross_val_score(i,df.drop('Churn',axis=1),y,scoring='accuracy',cv=5,n_jobs=-1,verbose=1)
    mns.append(cvs.mean())
    stds.append(cvs.std())
```

```python
for i in range(7):
    print(clfs[i].__class__.__name__,':',mns[i]*100)
```
